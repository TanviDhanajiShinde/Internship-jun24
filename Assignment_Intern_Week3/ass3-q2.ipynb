```python
def ol(data):
    q3=np.percentile(data,75)
    q1=np.percentile(data,25)
    iqr=q3-q1
    ub=q3+(1.5*iqr)
    lb=q1-(1.5*iqr)
    return ub,lb
```

```python
ub,lb=ol(df['Application mode'])
```

```python
# filtering the data without outliers
df=df[(df['Application mode']<= ub)&(df['Application mode']>=lb)]
```

```python
sns.boxplot(df['Application mode']);
plt.title('outliers in income');
```
