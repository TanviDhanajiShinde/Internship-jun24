```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([23,34,42,49,55,61,68,73,78,82])
y=np.array([25,29,33,37,44,47,57,62,67,71])
z=np.array([87,82,78,73,67,63,55,51,47,39])
```

#### Line Plot

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);
plt.grid()
# markers => >,<,^,H,o,*
```

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,z,marker='*',mfc='red',color='black',ls='--',ms=5);
plt.grid()
# markers => >,<,^,H,o,*
```

```python
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5,label='2023');
plt.plot(x,z,marker='*',mfc='red',color='black',ls='--',ms=5,label='2024');
plt.grid()
plt.legend();
# markers => >,<,^,H,o,*
```

```python
# Subplot
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);
plt.grid()
plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))

plt.plot(x,z,marker='*',mfc='red',color='black',ls='--',ms=5,label='2024');
plt.grid()
plt.legend();
```

### Bar Plot

```python
courses=['java','python','kotlin','dart','R']
students=[67,63,59,55,47]
```

```python

plt.title('Bar Plot');
plt.xlabel('Courses');
plt.ylabel('Students');
plt.bar(courses,students,color='Black',edgecolor='red',width=0.2,linewidth=3,alpha=0.5)
plt.grid()
```

#### Pie chart

```python
colors=['Black','Pink','Black','Pink','Black']
plt.pie(students,labels=courses,autopct='%.2f%%',labeldistance=1.1,startangle=90,shadow=True,explode=[0.3,0,0,0,0],colors=colors);   # starting angle default is 0
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.xticks(range(1,90,5))
plt.yticks(range(1,90,5))
plt.plot(x,y,marker='o',mfc='red',color='black',ls='--',ms=5);
plt.grid()
plt.subplot(2,2,3)
colors=['Black','Pink','Black','Pink','Black']
plt.pie(students,labels=courses,autopct='%.2f%%',labeldistance=1.1,startangle=90,shadow=True,explode=[0.3,0,0,0,0],colors=colors);
plt.subplot(2,2,2)
plt.scatter(df['Age'],df['EstimatedSalary'],c=df['Purchased']);
plt.subplot(2,2,4)
plt.title('Bar Plot');
plt.xlabel('Courses');
plt.ylabel('Students');
plt.bar(courses,students,color='Black',edgecolor='red',width=0.2,linewidth=3,alpha=0.5)
plt.grid()
plt.savefig('myplot.png')
```
